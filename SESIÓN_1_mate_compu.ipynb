## **Rango**
Estructura para almacenar sucesiones finitas de números enteros. Es inmutable!.
[Referencia](https://docs.python.org/es/3/library/stdtypes.html#tuple)

````python
sucesion = range(7)
print(type(sucesion))
print(sucesion)
print(sucesion[0])
print(sucesion[1])
print(sucesion[2])
print(sucesion[3])
print(sucesion[4])
len(sucesion)

print("El objeto sucesion tiene",len(sucesion),"elementos")
````

````python
rango = range(3,7) #aquí es cerrado en 3 y abierto en 7
print(rango)
print(rango[0])
print(rango[3]) #el elemento de índice 3, sería el 6
print(rango.stop) #el útimo elemento de la lista +1, osea sería 6+1=7
````

````python
rango = range(3,7,2) #la razón es la última entrada, osea la razón es 2. Osea estarían 3,5
print(rango)
print(rango[0])
print(rango[1]) #el elemento de índice 1,
print(rango.stop) #el útimo elemento de la lista +1, osea sería 6+1=7
````

## Conjunto (set)
Estructura de datos con almacenamiento y operaciones similar al de los conjuntos matemáticos.
[Referencia](https://www.w3schools.com/python/python_sets.asp)

````python
A1 = {"rojo", "verde", "azul"}
A2 = {"verde", "negro", "amarillo"}
print(type(A1))
print(A1)
````

````python
A3 = A1.intersection(A2) #intersección
print(type(A3))
print(A3)
A4 = A1.difference(A2) #diferencia
print(A4)
A5=A1.union(A2)
print(A5)
````

````python
A6=A1.isdisjoint(A2) #disjunción (osea hay o no intersección)
print(A6)
A7=A1.__contains__("ab") #¿está contenido?
print(A7)
````

````python
c=set(range(4,7))
print(type(c))
print(c)
````

````python
d=set([range(4,7),"hola",13])
print(type(d))
print(d)
````

````python
D={"Jorge","Pedro","Juan","Hector"} #para eliminar según orden de python
D.pop()
print(D)
D.remove("Pedro") #   para eliminar uno específico
print(D)
D.add(24) #para añadir
print(D)
````

````python
D={"Jorge","Pedro","Juan","Hector"}
for i in D: #i=iterador osea recorre las entradas uno por uno
    print(i)
````
